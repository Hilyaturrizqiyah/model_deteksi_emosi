```python
def remove_punct(text):
    text = re.sub(r'[^a-zA-Z]', ' ', str(text))
    text = re.sub(r'\b\w{1,2}\b', '', text)
    text = re.sub(r'\s\s+', ' ', text)
    text = re.sub('([#])|([^a-zA-Z])', ' ', text)
    text = re.sub(r'www.', '', text)
    text = re.sub(r'http\S+', '', text)
    return text


# df['twitter_account'] = df['username'].apply(lambda x: remove_punct(x))
df['ISI'] = df['Tweets'].apply(lambda x: remove_punct(x))
df.head()
```

```python
def str_to_lower(text):
    return text.lower()


# df['author'] = df['twitter_account'].apply(lambda x: str_to_lower(x))
df['sentiment'] = df['ISI'].apply(lambda x: str_to_lower(x))
df.head()
```

```python
def tokenization(text):
    # text = nltk.tokenize.word_tokenize(text)
    # return(text)
    text = re.split(r'\W+', text)
    return text


df['tokenization'] = df['sentiment'].apply(lambda x: tokenization(x))
df.head()
```

```python
stopword = pd.read_csv('stopwordbahasa.csv')
stopwords = [x for x in stopword['ada']]
stopwords
```

```python
def remove_stopword(text):
    text = [word for word in text if word not in stopwords]
    return text


df['remove_stopword'] = df['tokenization'].apply(lambda x: remove_stopword(x))
df.head()
```

```python
normalisasi_kata = pd.read_csv("kamus_singkatan.csv")

normalisasi_kata_dict = dict()

# for row in normalisasi_kata.iterrows():
#     if row[0] not in normalisasi_kata_dict:
#         normalisasi_kata_dict[row[0]] = row[1]


# def normalisasi_term(document):
#     return [normalisasi_kata_dict[term]
#             if term in normalisasi_kata_dict
#             else term for term in document]

for index, row in normalisasi_kata.iterrows():
    normalisasi_kata_dict[row['singkatan']] = row['normal']


def normalisasi_term(document):
    result = [normalisasi_kata_dict.get(w, w) for w in document]
    return result


df['normalisasi_tweet'] = df['remove_stopword'].apply(normalisasi_term)
df['normalisasi_tweet'].head(0)
```
